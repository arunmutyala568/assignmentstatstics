```python

"""A test is conducted which is consisting of 20 MCQs (multiple choices questions) with every
MCQ having its four options out of which only one is correct. Determine the probability that
a person undertaking that test has answered exactly 5 questions wrong."""






from math import factorial

from math import pow

from scipy.stats import binom

from scipy.stats import norm

import numpy as np

def prop(chances,success,currentpropability):
    
    binomialcofficient=factorial(chances)/(factorial(success)*factorial(chances-success))
    
    
    mypropability=binomialcofficient*(pow(currentpropability,success))*(pow(1-currentpropability,chances-success))
    
    
    print(mypropability)
    
    
prop(20,5,3/4)


data= binom.pmf(5, 20, 3/4)


print(prop)

print(data)
    
```

```python

"""Blood glucose levels for obese patients have a mean of 100 with a standard deviation of
15. A researcher thinks that a diet high in raw cornstarch will have a positive effect on
blood glucose levels. A sample of 36 patients who have tried the raw cornstarch diet
have a mean glucose level of 108. Test the hypothesis that the raw cornstarch had an
effect or not.""""


from scipy.stats import norm

import numpy as np
def hypo(nullhypo,alternatehypo,n,sigma):
    
    
    zalpha=0.05
    
    criticalvalues=norm.ppf(zalpha/2)
    
    print(criticalvalues)
    
    zvalue=(alternatehypo-nullhypo)/(sigma/(np.sqrt(n)))
    print(zvalue)   
    
    if zvalue<abs(criticalvalues)*-1 or zvalue>abs(criticalvalues):
        
        print("in rejection region , rejecting null hypthesis")
        
        
    else:
        
        print("inside confidence level, reject alternate")
        
        
        

        
        

hypo(100,105,75,15)




    
```
